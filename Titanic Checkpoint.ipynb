```python
#Après avoir importé l' ensemble de données à l'aide de Pandas, commencez par afficher l'en-tête de l'ensemble de données, puis quelques informations générales sur les colonnes de données et les valeurs.

import pandas as pd

df=pd.read_csv('titanic-passengers.csv', sep=";" )
print(df.head())

print(df.info())

```

```python
#Dans une prochaine étape, appliquez ce que vous avez appris dans le prétraitement des données pour prétraiter vos données : trouvez les informations manquantes et remplacez-les par les valeurs appropriées.

print(df.isnull().sum())
```

```python
#remplacer les valeurs manquantes par les valeurs appropriées.

#Pour la variable AGE ON UTILISE LA MOYENNE
df['Age'].fillna(df['Age'].mean(), inplace=True)
print(df.isnull().sum())
```

```python
#Pour la variable Cabine on utilise le mode

df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
print(df.isnull().sum())
```

```python
#Pour la variable Embarked on utilise le mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print(df.isnull().sum())
```

```python
#Pour la variable Embarked on utilise le mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print(df.isnull().sum())
```
