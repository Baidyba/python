Dans ce point de contrôle, nous allons prédire si un patient a une maladie rénale chronique ou non en utilisant plusieurs algorithmes de classification d'apprentissage supervisé.

La maladie rénale chronique est une diminution du fonctionnement des reins qui ne filtrent plus correctement le sang de l'organisme. Cette insuffisance rénale chronique a deux causes principales : le diabète et l'hypertension artérielle.

# Partie 1 : apprentissage supervisé

```python
# Chargez l' ensemble de données de la maladie rénale.

import pandas as pd

data=pd.read_csv('kidney_disease.csv')
```

```python
# Utilisez cette commande pour renommer les colonnes

data.rename(columns={'bp': 'blood_pressure', 'sg': 'specific gravity','al':'albumin','su':'sugar','rbc ':'globules rouges','pc':'cellule de pus','pcc':'amas de cellules de pus','ba':'bactéries','bgr':'glycémie aléatoire','bu':'urée sanguine','sc':'créatinine sérique','sod':'sodium','pot':'potassium','hemo':'hémoglobine','pcv':'hématocrite','wc' :'nombre de globules blancs','rc':'nombre de globules rouges','htn':'hypertension','dm':'diabète sucré','cad':'maladie coronarienne','appet': 'appétit','pe':'pédale œdème','ane':'anémie','classification':'classe'},inplace=True)
```

```python
data['age'].fillna(data['age'].mean(), inplace=True)
data['blood_pressure'].fillna(data['blood_pressure'].mean(), inplace=True)
data['specific gravity'].fillna(data['specific gravity'].mean(), inplace=True)
data['albumin'].fillna(data['albumin'].mean(), inplace=True)
data['sugar'].fillna(data['sugar'].mean(), inplace=True)
data['glycémie aléatoire'].fillna(data['glycémie aléatoire'].mean(), inplace=True)
data['urée sanguine'].fillna(data['urée sanguine'].mean(), inplace=True)
data['créatinine sérique'].fillna(data['créatinine sérique'].mean(), inplace=True)
data['sodium'].fillna(data['sodium'].mean(), inplace=True)
data['potassium'].fillna(data['potassium'].mean(), inplace=True)
data['hémoglobine'].fillna(data['hémoglobine'].mean(), inplace=True)
                    
data['rbc'].fillna(data['rbc'].mode()[0], inplace=True)
data['cellule de pus'].fillna(data['cellule de pus'].mode()[0], inplace=True)
data['amas de cellules de pus'].fillna(data['amas de cellules de pus'].mode()[0], inplace=True)
data['bactéries'].fillna(data['bactéries'].mode()[0], inplace=True)
data['hématocrite'].fillna(data['hématocrite'].mode()[0], inplace=True)
data['nombre de globules blancs'].fillna(data['nombre de globules blancs'].mode()[0], inplace=True)
data['nombre de globules rouges'].fillna(data['nombre de globules rouges'].mode()[0], inplace=True)
data['hypertension'].fillna(data['hypertension'].mode()[0], inplace=True)
data['diabète sucré'].fillna(data['diabète sucré'].mode()[0], inplace=True)
data['maladie coronarienne'].fillna(data['maladie coronarienne'].mode()[0], inplace=True)
data['appétit'].fillna(data['appétit'].mode()[0], inplace=True)
data['pédale œdème'].fillna(data['pédale œdème'].mode()[0], inplace=True)
data['anémie'].fillna(data['anémie'].mode()[0], inplace=True)
data['classe'].fillna(data['classe'].mode()[0], inplace=True)

```
