```python
import pandas as pd

url= 'https://api.covidactnow.org/v2/states.csv?apiKey=89e6ecfa2d144ba49bef6f885aa75f66'
df = pd.read_csv(url)
df
```

```python
## Garder les élément dans une variable datacv

datacv=df[['population','metrics.testPositivityRatio','metrics.caseDensity','metrics.infectionRate','metrics.bedsWithCovidPatientsRatio','actuals.hospitalBeds.weeklyCovidAdmissions','actuals.hospitalBeds.weeklyCovidAdmissions','metrics.weeklyCovidAdmissionsPer100k','communityLevels.canCommunityLevel','actuals.vaccinationsFall2022BivalentBooster','metrics.vaccinationsFall2022BivalentBoosterRatio','actuals.positiveTests']]
```

```python
##supprimer les colonnes qui n'ont pas de nombre
datacv=datacv.drop(['metrics.caseDensity','metrics.infectionRate'], axis=1)
```

```python
##regarder les valeurs manquantes de notre tableau
datacv.isnull().sum()
```

```python
## on remplace les valeus manquante par la moyenne
datacv['metrics.testPositivityRatio'].fillna(datacv['metrics.testPositivityRatio'].mean(), inplace=True)
datacv['metrics.bedsWithCovidPatientsRatio'].fillna(datacv['metrics.bedsWithCovidPatientsRatio'].mean(), inplace=True)
datacv['actuals.hospitalBeds.weeklyCovidAdmissions'].fillna(datacv['actuals.hospitalBeds.weeklyCovidAdmissions'].mean(), inplace=True)
datacv['metrics.weeklyCovidAdmissionsPer100k'].fillna(datacv['metrics.weeklyCovidAdmissionsPer100k'].mean(), inplace=True)
datacv['communityLevels.canCommunityLevel'].fillna(datacv['communityLevels.canCommunityLevel'].mean(), inplace=True)
datacv['actuals.vaccinationsFall2022BivalentBooster'].fillna(datacv['actuals.vaccinationsFall2022BivalentBooster'].mean(), inplace=True)
datacv['metrics.vaccinationsFall2022BivalentBoosterRatio'].fillna(datacv['metrics.vaccinationsFall2022BivalentBoosterRatio'].mean(), inplace=True)
```

```python
datacv.head()
```

```python
datacv = datacv.drop_duplicates()
```

```python
datacv.isnull().sum()
```

```python
## La regession multilinéaire

x=datacv[['metrics.testPositivityRatio','metrics.bedsWithCovidPatientsRatio','metrics.weeklyCovidAdmissionsPer100k','communityLevels.canCommunityLevel','actuals.vaccinationsFall2022BivalentBooster','metrics.vaccinationsFall2022BivalentBoosterRatio']]
y=datacv['actuals.positiveTests'].values
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40) 
model=LinearRegression()
model.fit(x_train,y_train) 
predicted=model.predict(x_test) 
```

```python
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

```python
print("Accuracy={:.2f}".format(model.score(x_test, y_test)))
```
