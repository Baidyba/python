```python
## Importation des bibliothèque

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
### On définit X
X=data.drop(['job','marital','education','default','housing','loan','month','contact','campaign','poutcome','deposit'], axis=1)
```

```python
X
```

```python
##Remplacer par no et yes les valeurs 0 et 1 
data['deposit'].replace(['no','yes'],[0,1], inplace=True)
```

```python
Y=data['deposit']
```

```python
from sklearn.model_selection import train_test_split
num_classes = 2
batch_size = 128
epochs = 40

# Divisez les données en ensembles d'entraînement et de test (par exemple, 80% pour l'entraînement, 20% pour les tests)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# convert class vectors to binary class matrices
train_labels_categ = keras.utils.to_categorical(Y_train, 2)
test_labels_categ = keras.utils.to_categorical(Y_test, 2)
```

```python
train_labels_categ
```

```python
Y_train
```

```python
# Créez un modèle séquentiel
model = Sequential()

```

```python
# Ajoutez une couche d'entrée avec la taille appropriée (selon vos données)
model.add(Dense(64, input_shape=(6,), activation='relu'))

```

```python
# Ajoutez une couche cachée
model.add(Dense(128, activation='relu'))
```

```python
# Ajoutez une couche de sortie avec le nombre approprié de classes (selon vos données)
model.add(Dense(num_classes, activation='softmax'))
```

```python
# Compilez le modèle avec une fonction de perte et un optimiseur appropriés
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, train_labels_categ,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(X_test, test_labels_categ))
```

```python
score = model.evaluate(X_test, test_labels_categ, verbose=0)
print('Test accuracy:', score[1])
```
